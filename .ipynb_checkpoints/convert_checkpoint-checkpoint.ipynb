```python
# Geneformer-V1
# = BertForMaskedLM
# = vocab 25426
# = hidden 256
# = max length 2048
# = intermediate 512
# = likely 6 layers


# Geneformer-V2-104M
# = BertForMaskedLM
# = vocab 20275
# = hidden 768
# = max length 4096
# = intermediate 3072
# = likely 12 layers


# Geneformer-V2-104M_CLcancer
# = BertForMaskedLM
# = vocab 20275
# = hidden 768
# = max length 4096
# = intermediate 3072
# = likely 12 layers


# Geneformer-V2-316M
# = BertForMaskedLM
# = vocab 20275
# = hidden 1152
# = max length 4096
# = intermediate 4608
# = likely 18 layers


import torch
from pathlib import Path

ckpt_path = Path("./assets/Geneformer-V2-104M_CLcancer/pretrained_GeneformerV2-104M_CLcancer.pth")

ckpt = torch.load(ckpt_path, map_location="cpu")

print("checkpoint type:", type(ckpt))

if isinstance(ckpt, dict):
    print("top-level keys:", ckpt.keys())

    if "model_state_dict" in ckpt:
        state = ckpt["model_state_dict"]
        print("using: ckpt['model_state_dict']")
    elif "state_dict" in ckpt:
        state = ckpt["state_dict"]
        print("using: ckpt['state_dict']")
    else:
        state = ckpt
        print("using checkpoint directly as state_dict")
else:
    state = ckpt
    print("using checkpoint directly")

print("num tensors:", len(state))

for i, (k, v) in enumerate(state.items()):
    print(k, tuple(v.shape), v.dtype)
    if i >= 30:
        break
```

```python
# Pythonic way to run nano-Geneformer
import torch
import numpy as np
from time import time

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_name = "Geneformer-V2-316M" #Geneformer-V1 || Geneformer-V2-104M 
                                  #|| Geneformer-V2-104M_CLcancer || Geneformer-V2-316M

from model import GeneformerForMaskedLM
model = GeneformerForMaskedLM.from_pretrained(model_name).to(device)
model.eval()


start_time = time()
embs = model.encode(
    "./pancreas.h5ad",
    model_name=model_name,
    batch_size=256,
)

embs = embs.numpy()

print("embeddings:", embs.shape)

np.save("geneformer_v2_316M_embeddings.npy", embs)
total_time = time() - start_time
print(total_time)
```
